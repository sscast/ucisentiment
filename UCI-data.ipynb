The dataset
------

- This dataset contains 3000 reviews with scores
- Score is either 1 (for positive) or 0 (for negative)
- The sentences come from three different websites/fields:
    1. imdb.com
    2. amazon.com
    3. yelp.com
- For each website, there exists 1000 reviews, 500 positive and 500 negative. Those were selected randomly from larger datasets of reviews.
The reviews were selected to have a clearly positive or negative connotaton, with a goal that no neutral sentences to be selected.
You may download the data at this [link](http://archive.ics.uci.edu/ml/datasets/Sentiment+Labelled+Sentences). 

Considerations
------
- The dataset has been selected with umambiguous label so it's perfect to do model training; however, the data size is not big so data might not be generalized enough and thus, may not be a good dataset for deep learning models. However, it still worth to use it as an experiment to test our understandings about different models
- The dataset is relatively clean: no need to deal with outliers, missing or misleading data, and analysis can be focused on model selection and classification

What to improve for this analysis
-----
- Using a larger dataset for training
- Add some wrapper classes, so the analysis can be automated 
- Hyper-parameter tuning can be automated
- Implementation is not optimized and not very re-usable at current state

## process data
### methods 

```python
import re

from PIL import Image

def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
#     text = re.sub()
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    #remove stopwords 
    text = ' '.join([word for word in text.split() if word not in stopwords])
    return text


# get top n grams
def get_top_n_grams(corpus, n_gram_value, n=None):
    
    vector = CountVectorizer(ngram_range=(n_gram_value, n_gram_value), stop_words=stopwords).fit(corpus)
    
    # bag of words representation
    bow = vector.transform(corpus)
    sum_words = bow.sum(axis=0)
    
    # Determine frequency for the chart
    words_freq = [(word, sum_words[0, idx]) for word, idx in vector.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    
    return words_freq[:n]

def get_combined_top_n_grams(top_n_1, top_n_0): 
    top_n_1=pd.DataFrame(top_n_1,columns=['word','count'])
    top_n_1['label']=1
    top_n_0 =pd.DataFrame(top_n_0,columns=['word','count'])
    top_n_0['label']=0
    top_n = pd.concat([top_n_1, top_n_0])
    return top_n

def plot_n_gram(top_n_grams):
    plt.subplots_adjust(left=None, bottom=None, right=None, top= 0.8, wspace=0.3 , hspace=0.3)
    g=sns.catplot(x="count", y="word", col="label", data=top_n_grams, orient="h", saturation=.5, kind="bar", sharex=False, sharey=False)
    g.set(xlim=(0, 110), ylim=(0, 20))
    (g.set_axis_labels("Word count", "Count").despine(left=True))  
```

### Classification
#### Tokenization 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
import pickle

from sklearn.model_selection import train_test_split



def create_train_test_data(X, y, div, fileName):
    # X, y are two columns from pandas dataframe 

    # set the min frequency to 2 
    vectorizer = TfidfVectorizer(ngram_range=(1, 3), analyzer='word', min_df=2)
    

    TOP_K=15000 # set a max number to avoid too many features
        

    X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=div)
    
    # vectorize data
    X_train = vectorizer.fit_transform(X_train)
    
    X_test = vectorizer.transform(X_test)

    selector = SelectKBest(f_classif, k=min(TOP_K, X_train.shape[1]))

    selector.fit(X_train, y_train)
    X_train = selector.transform(X_train).astype('float32')
    X_test = selector.transform(X_test).astype('float32')
      
    # save the vectorizer
    with open('pickle/'+fileName +'.pickle', 'wb') as handle:
        pickle.dump(vectorizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
    return X_train, y_train, X_test, y_test
```

##### Logistic Regression

```python
from sklearn import linear_model, naive_bayes, metrics

def train_model(classifier, feature_vector_train, label, feature_vector_valid, ytest):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, label)
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)
    
    report = metrics.classification_report(ytest, predictions)
    accuracy = metrics.accuracy_score(predictions, ytest)
    return accuracy, report, classifier
```

##### Simple MLP model
> Simple logistic regression and naive bayes seem quite good, but there's room to improve, I decided to approach a nueral network model to compare. Making reference to Google's Keras guide, for small datasets, the best model is to use a Simple Multilayer Perceptron (MLP) model, because the dataset won't benefit from a sequence model. I copied the following piece of code, which provides an wrapper to use the model easily, with the freedom to add a few dropout layers for regularization. <br><br>
> In the implementation, I have used both unigram and bigram tokens to feed and train the model, based on the clustering analysis in the previous section, as bigram tokens have some important information for our model too and I don't want to lose that. <br><br>
> I will use relu activation function and the network will be trained using the efficient Adam implementation of gradient descent and the binary cross entropy loss function, suited to binary classification problems. <br> <br>
> Output layer is a single neuron with a sigmoid activation function for predicting 0 for negative and 1 for positive reviews

```python
def mlp_model(layers, units, dropout_rate, input_shape, num_classes):
    """Creates an instance of a multi-layer perceptron model.

    # Arguments
        layers: int, number of `Dense` layers in the model.
        units: int, output dimension of the layers.
        dropout_rate: float, percentage of input to drop at Dropout layers.
        input_shape: tuple, shape of input to the model.
        num_classes: int, number of output classes.

    # Returns
        An MLP model instance.
    """
    op_units, op_activation = _get_last_layer_units_and_activation(num_classes)
    model = Sequential()
    model.add(Dropout(rate=dropout_rate, input_shape=input_shape))

    for _ in range(layers-1):
        model.add(Dense(units=units, activation='relu'))
        model.add(Dropout(rate=dropout_rate))

    model.add(Dense(units=op_units, activation=op_activation))
    
    return model

def _get_last_layer_units_and_activation(num_classes):
    """Gets the # units and activation function for the last network layer.

    # Arguments
        num_classes: int, number of classes.

    # Returns
        units, activation values.
    """
    if num_classes == 2:
        activation = 'sigmoid'
        units = 1
    else:
        activation = 'softmax'
        units = num_classes
    return units, activation
```

```python
layers = 1 # add dropout layers for regularization 
units = 32 
dropout_rate = 0.01
input_shape = Xtrain.shape[1:]
num_classes = 2
num_epochs=120
learning_rate=0.001


model = mlp_model(layers, units, dropout_rate, input_shape, num_classes)

optimizer = tf.keras.optimizers.Adam(lr=learning_rate)
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
